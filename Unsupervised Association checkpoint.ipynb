```python
import pandas as pd 
dataset1=pd.read_csv("Market_Basket_Optimisation.csv", sep=",",header= None)
dataset1.info()
dataset=[]
for i in range(0,7001):
    
    dataset.append([str(dataset1.values[i,j]) for j in range(0,20)])
    
```

```python
import mlxtend
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
te=TransactionEncoder()
te_ary=te.fit(dataset).transform(dataset)    #Apply one-hot-encoding on our dataset
df=pd.DataFrame(te_ary, columns=te.columns_)  #Creating a new DataFrame from our Numpy array
df
```

```python
for c in df.columns:
    print(c)
```

```python
df = df.drop('nan',axis=1)
```

```python
for c in df.columns:
    print(c)
```

```python
from mlxtend.frequent_patterns import apriori
apriori(df, min_support=0.1)
```

```python
frequent_itemsets=apriori(df, min_support=0.04, use_colnames=True) #Instead of column indices we can use column names.
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.1)
```

```python
from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets,metric="lift",min_threshold=1.25)
```
