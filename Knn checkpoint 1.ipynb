```python
import math

def euclideanDistance(instance1, instance2, length):
    somme = 0
    for i in range(length):
        somme += math.pow(instance1[i]-instance2[i],2)
        
    distance =math.sqrt(somme )
    return distance

data1 = [2, 2, 2, 'a']

data2 = [4, 4, 4, 'b']

distance = euclideanDistance(data1, data2, 3)
print ('Distance: ' + repr(distance))
print ('Distance: ' + str(distance))
```

```python
import operator

def getNeighbors(trainingSet, testInstance, k):

    distances = []

    length = len(testInstance)-1

    for x in range(len(trainingSet)):

        dist = euclideanDistance(testInstance, trainingSet[x], length)

        distances.append((trainingSet[x], dist))

        distances.sort(key=operator.itemgetter(1))

    neighbors = []

    for x in range(k):

        neighbors.append(distances[x][0])

    return neighbors


trainSet = [[2, 2, 2, 'a'], [4, 4, 4, 'b']]

testInstance = [5, 5, 5]

k = 1

neighbors = getNeighbors(trainSet, testInstance, 1)

print(neighbors)
```

```python
def getAccuracy(testSet, predictions):
    correct = 0
    #Complete the function
    for i in range (len(predictions)):
        if predictions[i] == testSet[i][3]:
            correct += 1

    return (correct/float(len(testSet))) * 100.0



testSet = [[1,1,1,'a'], [2,2,2,'a'], [3,3,3,'b']]

predictions = ['a', 'a', 'a']

accuracy = getAccuracy(testSet, predictions)

print(accuracy)
```

```python
import math

def manhattanDistance(instance1, instance2, length):
    somme = 0
    for i in range(length):
        y= instance1[i]-instance2[i]
        if y < 0:
            y = y * -1
        somme += y
        
    distance =somme 
    return distance

data1 = [2, 2, 2, 'a']

data2 = [4, 4, 4, 'b']

distance = manhattanDistance(data1, data2, 3)
print ('Distance: ' + repr(distance))
print ('Distance: ' + str(distance))
```

```python
import math

def minkowskiDistance(instance1, instance2, length,p):
    somme = 0
    for i in range(length):
        y= instance1[i]-instance2[i]
        if y < 0:
            y = y * -1
        somme += math.pow(y,p)
        
    distance =math.pow(somme, 1 / p )
    return distance

data1 = [2, 2, 2, 'a']

data2 = [4, 4, 4, 'b']
p=1
distance = minkowskiDistance(data1, data2, 3,p)
print ('Distance Manhattan  p = 1  : ' + repr(distance))
p=2
distance = minkowskiDistance(data1, data2, 3,p)
print ('Distance Euclidean  p = 2  : ' + str(distance))
p=3
distance = minkowskiDistance(data1, data2, 3,p)
print ('Distance   p = 3  : ' + str(distance))
p=4
distance = minkowskiDistance(data1, data2, 3,p)
print ('Distance   p = 4  : ' + str(distance))
```
